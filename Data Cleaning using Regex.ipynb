```python
import pandas as pd
import re
df = pd.read_csv("datasets/presidents.csv")
df.head()
```

```python
df['FirstName'] = df['President'].str.extract('(\w+)')
df['MiddleName'] = df['President'].str.extract('(?:[\w]+\s)(\w+[.]?)(?:\s\w+)')
df['LastName'] = df['President'].str.extract('([\w-]+$)')
df[['President','FirstName','MiddleName','LastName']]
```

```python
df['Born'] = df['Born'].str.extract('(\w{1,3}\s\d{1,2}[,]\s\d{1,4})')
df['Born'] = pd.to_datetime(df['Born'])
df
```

```python
df['Died'] = df['Died'].str.extract('(\w{1,3}\s\d{1,2}[,]\s\d{1,4})')
df['Died'] = pd.to_datetime(df['Died'])
```

```python
df['Start_Year_of_Presidency'] = df['Age atstart of presidency'].str.extract('(\d{1,2})(?:\s)(?:\w{1,5})(?:[,])(?:\s)(?:\d{1,3})(?:\s\w{1,4})(?:\w{1,3}\s\d{1,2}[,]\s\d{1,4})')
df['Start_Days_of_Presidency'] = df['Age atstart of presidency'].str.extract('(?:\d{1,2})(?:\s)(?:\w{1,5})(?:[,])(?:\s)(\d{1,3})(?:\s\w{1,4})(?:\w{1,3}\s\d{1,2}[,]\s\d{1,4})')
df['Start_Date_of_Presidency'] = df['Age atstart of presidency'].str.extract('(?:\d{1,2})(?:\s)(?:\w{1,5})(?:[,])(?:\s)(?:\d{1,3})(?:\s\w{1,4})(\w{1,3}\s\d{1,2}[,]\s\d{1,4})')
df['Start_Date_of_Presidency'] = pd.to_datetime(df['Start_Date_of_Presidency'])
df
```

```python
df['End_Year_of_Presidency'] = df['Age atend of presidency'].str.extract('(\d{1,2})(?:\s)(?:\w{1,5})(?:[,])(?:\s)(?:\d{1,3})(?:\s\w{1,4})(?:\w{1,3}\s\d{1,2}[,]\s\d{1,4})')
df['End_Days_of_Presidency'] = df['Age atend of presidency'].str.extract('(?:\d{1,2})(?:\s)(?:\w{1,5})(?:[,])(?:\s)(\d{1,3})(?:\s\w{1,4})(?:\w{1,3}\s\d{1,2}[,]\s\d{1,4})')
df['End_Date_of_Presidency'] = df['Age atend of presidency'].str.extract('(?:\d{1,2})(?:\s)(?:\w{1,5})(?:[,])(?:\s)(?:\d{1,3})(?:\s\w{1,4})(\w{1,3}\s\d{1,2}[,]\s\d{1,4})')
df['End_Date_of_Presidency'] = pd.to_datetime(df['End_Date_of_Presidency'])
df
```

```python
df['years_Post-presidencytimespan'] = df['Post-presidencytimespan'].str.extract('(\d{1,2})(?:\s)(?:\w{1,5})(?:[,])(?:\s)(?:\d{1,3})(?:\s\w{1,4})')
df['Days_Post-presidencytimespan'] = df['Post-presidencytimespan'].str.extract('(?:\d{1,2})(?:\s)(?:\w{1,5})(?:[,])(?:\s)(\d{1,3})(?:\s\w{1,4})')
df
```

```python
df['years_Age'] = df['Age'].str.extract('(\d{1,2})(?:\s)(?:\w{1,5})(?:[,])(?:\s)(?:\d{1,3})(?:\s\w{1,4})')
df['Days_Age'] = df['Age'].str.extract('(?:\d{1,2})(?:\s)(?:\w{1,5})(?:[,])(?:\s)(\d{1,3})(?:\s\w{1,4})')
df
```
